**Finding file of largest size**

we are going to find file large then 1 MB in directory 'os-test' hirerchical structure

```python
# we are getting the size of all the files or folder in 'os-test' directory
file_sizes = []

# "os.walk" is a function that do traversal on all the directory tree and even of file in directory folder

for dir_path, dirs, files in os.walk("."):      # "." represents current directory

    # "dirs" give the directorys(folder) present in the current directory traversal

    print("Running in: ", dir_path)     # "dir_path" give you the current directory where it is traversal
    
    for f in files:             # "files" give you files present in current directory

        file_full_path = os.path.join(dir_path, f)      # making path for each file to tell where it is present
        file_size = os.path.getsize(file_full_path)     # telling the size of each file
        
        # we are making a list of tuples in which first element represent path of file where as second element tell the file size 
        file_sizes.append((file_full_path, file_size))      
        
    print("")
```

```python
import pprint
pprint.pprint(file_sizes)
```

```python
file_sizes.sort(key=lambda x: x[1], reverse=True)       # we are sorting by the size (which is second element of tuple)
```

```python
file_sizes
```

```python
filtered = filter(lambda x:x[1]> 1024*1024, file_sizes)         # get files greater than 1 MB(1024*1024)
for f in filtered:
    print(f)
```

# **Cards and Decks**

```python
class Card:
    def __init__(self, suit, val):
        self.suit = suit
        self.val = val
        
    def __str__(self):
        return str(self.val) + " of " + self.suit
    
class Deck:
    def __init__(self):
        self.cards = []
        self.build()
        
    def build(self):
        for s in ["Spades", "Hearts", "Diamonds", "Clubs"]:
            for v in range(1, 14):
                self.cards.append((Card(s, v)))
                
                
    def __str__(self):
        ret = ""
        for c in self.cards:
            ret += str(c) + "\n"
        return ret
```

```python
c1 = Card("Hearts", 5)
```

```python
print(c1)
```

```python
d = Deck()
```

```python
print(d)
```

```python
import random
```

```python
def shuffle(self):
    for i in range(0, len(self.cards)):
        r = random.randint(0, i)        # find another number
        self.cards[i], self.cards[r] = self.cards[r], self.cards[i]      # and swap
        
Deck.shuffle = shuffle
```

```python
d.shuffle()
print(d)
```

```python
def draw(self):
    r = random.randint(0, len(self.cards))
    c = self.cards.pop(r)
    return c

Deck.draw = draw
```

```python
c = d.draw()
```

```python
print(c)
```

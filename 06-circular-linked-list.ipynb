# **Circular Linked List(or Ring Data Structure)**

This is also similar to a regular linked list -- expect the last node is "linked" to the head thus creating a ring .

Due to this we can't "loop untill None is reached" because we will never have a None in the next now-- even for a single node! So, we need to keep track of the head and always loop "untill we reach back to head".

```python
class Node:
    def __init__(self, data = None):
        self.val = data
        self.next = None

class Ring: 
    def __init__(self):
        self.head = None
```

Let's explain this reasoning through trying to output the ring.

```python
def __str__(self):
    ret_str = '['
    temp = self.head
    while temp is not None:
        ret_str += str(temp.val) + ', '
        temp = temp.next
        
        if temp == self.head:       # different for ring
            break
        
    ret_str = ret_str.rstrip(', ')
    ret_str += ']'
    
    return ret_str

Ring.__str__ = __str__
```

```python
def _get_last(self):     # helper funtion
    # no node, no last
    if self.head is None:
        return None
        
    # by using the below code you not need to solve case 2 (single node)   
    temp = self.head
    while temp.next != self.head:
        temp = temp.next
    return temp

Ring._get_last = _get_last
```

```python
def insert(self, index, val):
    new_node = Node(val)
    
    last = self._get_last()     # need last for ring (change)
    
    # insertion at index 0 is different
    if index == 0:
        
        new_node.next = self.head
        self.head = new_node
        
        # also need to set the last pointer to this new head(change)
        if last is None:
            self.head.next = self.head # first node ever being inserted
        else:
            last.next = new_node
            
            
        return
    
    if self.head is None:   # and index > 0:
        raise IndexError("Cannot insert at "+ str(index) + " because list is empty")
    # for other indices 
    temp = self.head
    
    counter = 0
    while temp is not None and counter < index:         # temp will never be None! 
        prev = temp
        temp = temp.next
        counter += 1
        
    prev.next = new_node
    new_node.next = temp
    
Ring.insert = insert
```

```python
r = Ring()
r.insert(0,1)
r.insert(0,2)
r.insert(1,3)
r.insert(7,5)     # difference behaviour since it's a ring !
print(r)
```

```python
def remove(self, val):
    # empty ring case
    if self.head is None:
        return
    
    
    temp = self.head
    last = self._get_last()     # (change)
    
    # first node matches case   # (change)
    if temp.val == val:
        if last == self.head:
            self.head = None    # just one node. Now gone
            
        else:
            self.head = temp.next
            last.next = self.head
        return
    
    # let's move to next nodes
    # temp holds the value of the node that will be deleted
    prev = temp         # (change)
    temp = temp.next    # (change)
    
    while temp != self.head:    # (change)
        if temp.val == val:
            break
        
        prev = temp 
        temp = temp.next
        
    if temp == self.head:    # not found
        return
        
    prev.next = temp.next   # just lose the reference to delete node
    
Ring.remove = remove
```

```python
r = Ring()
r.insert(0,1)
r.insert(0,2)
r.insert(1,3)
r.insert(7,5)     # difference behaviour since it's a ring !
print(r)
r.remove(1)
print(r)
```

```python
def len(self):
    if self.head is None:
        print("Empty List")
        return
    counter= 1
    last = self.head
    while last.next is not None and last.next is not self.head:
        counter +=1
        last = last.next
    return counter
Ring.len = len
```

```python
d = Ring()
d.len()
```

```python
def get(self, index):
    if  self.head == None:
        return None
    if index == 0:
        return self.head.val
    temp = self.head.next
    counter = 0 
    while temp is not self.head:
        counter +=1
        if counter == index:
            break
        temp = temp.next
    if temp == self.head:
        return None
    return temp.val
    
Ring.get = get
```

```python
print(d.get(0))
```

```python
def push(self, val):
    new_node = Node(val)
    if self.head is None:
        self.head = new_node
        new_node.next = self.head
        return
    
    last = _get_last(self)
    last.next = new_node
    new_node.next = self.head
    
    
Ring.push = push
        
```

```python
r.push(1)
print(r)
d.push(2)
d.push(1)
print(d)
```

```python
def pop(self):
    if self.head is None:
        raise Exception("Cannot pop, NO value ")
    last = self.head
    if last.next == self.head:
        self.head = None
        return
    while last.next != self.head:
        prev = last
        last = last.next
    prev.next = last.next

Ring.pop = pop
```

```python
print(d)
d.pop()
print(d)
d.pop()
print(d)
```

```python
def removeAt(self, index):
    last =_get_last(self)
    if self.head is None:
        raise Exception("List is Empty")
    temp = self.head
    if index == 0:
        if temp.next == self.head:
            self.head = None
        else:
            self.head = self.head.next
            last.next = self.head
        return
    counter = 0 
    prev = None
    while temp.next != self.head:   
        if counter == index:
            break
        prev = temp
        temp = temp.next
        counter += 1
        
    if temp == self.head or counter < index:
        raise Exception("OUT OF RAnge")
    prev.next = temp.next
Ring.removeAt = removeAt
```

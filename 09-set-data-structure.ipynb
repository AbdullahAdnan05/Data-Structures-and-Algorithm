**All and Any**

You can all and any  if you have a lot of condition -- all or some of which need to be satisfied.

```python
a = 5
b = 4

if a > 2 and a < 10 and b > 0 and b < 10:
    print("All constraints satisfied") 
```

```python
a = 5
b = 4
constraints = { a > 2 ,
                a < 10,
                b > 0 ,
                b < 10 }
```

```python
if all(constraints):        # all is keyword by which all elements in set is check one by one and it return true if all elements are true
    print("All constraints satisfied")
```

```python
if any(constraints):        # it will give true if anyone of given condition is true it give true
    print("At least one constraints satisfied")
```

**Quantifiers**

```python
X = {10,20,30,40,50}
```

**∀ x ∈ X , x < 100**

```python
all([x < 100 for x in X ] )     # same as "for all/each"
```

**∃ x ∈ X , x < 100**

```python
any ( [ x < 50 for x in X ] )       # same as "there exsists"
```

You can also have an arbitrary function computed in the scope.

```python
def is_div_by_20(n):
    return n%20 ==0
```

**∀ x ∈ X , is_div_by_20(x)**

```python
all( [ is_div_by_20(x) for x in X ] )
```

**∃ x ∈ X , is_div_by_20(x)**

```python
any( [ is_div_by_20(x) for x in X ] )
```

**Counter**

A Counter is the same as a dictionary except it is used to keep count and as such does not throw a KeyError

```python
from collections import Counter
cnt = Counter()
for word in ['red', 'blue', 'red', 'green', 'blue', 'blue']:
    cnt[word] += 1
```

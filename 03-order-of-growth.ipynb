**Order of Growth**

Recall fibonacci series:

1,1,2,3,5,8,13,...

fib(n) = fib(n-1) + fib (n-2)

We had two alternative for computing fib(n)

```python
def fib1(n):
    if n <= 1:
        return 1
    
    else:
        return fib1(n-2) + fib1(n-1)
```

```python
def fib2(n):
    a = 1
    b = 1
    
    for i in range(n):
        a, b = b, a+b
        
    return a
```

```python
print('fib: ')
for i in range(15):
    print(fib1(i), end = ", ")
    
print('\n\n')
print('fib2: ')
for i in range(15):
    print(fib2(i), end= ", ")
```

```python
import time

def compute_times(fn, limit):
    l = []
    
    for i in range (limit):
        
        start_time = int(round(time.time() * 1000))
        
        fn(i)
        
        end_time = int(round(time.time() * 1000))
        time_taken = end_time - start_time
        
        l.append(time_taken)
        
    return l
```

```python
limit = 30 
fib1_times = compute_times(fib1, limit)
fib2_times = compute_times(fib2, limit)
```

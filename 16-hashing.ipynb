# Hash Tables

```python
class HashMap:
    def __init__(self):
        self.size = 10
        self.map = [None] * self.size       # we are intializing the whole ten elements of list with none        
```

```python
def _get_hash(self, key):
    return key % self.size

HashMap._get_hash = _get_hash
```

```python
def add(self, key, value ):
    key_hash = self._get_hash(key)
    key_value = [key, value]
    
    
    # insert or update:  "upsert"
    
    self.map[key_hash] = [ key_value ]      # notice the double list. We'll get to it 
    return True         # it tell that you have insert it correctly
HashMap.add = add
```

```python
def get(self, key):
    key_hash = self._get_hash(key)
    if self.map[key_hash] is not None:
        for pair in self.map[key_hash]:
            if pair[0] == key:
                return pair[1]
    
    raise KeyError(str(key))

HashMap.get = get
```

```python
l = ['a', 'b', 'c']
for i, item in enumerate(l):        # "enumerate" give you two thing first index and second value at that index
    print(i,"---",item)         # i store in index where item store value
```

```python
def __str__(self):
    ret = ""
    
    for i, item in enumerate(self.map):     # i represents the index where as item represent value
        if item is not None:
            ret += str(i) + ": " + str(item) + "\n"
    return ret

HashMap.__str__ = __str__
```

## Collision and Avoidance

```python
h = HashMap()

h.add(17, "seventeen")
h.add(26, "twenty six")
h.add(35, "thirty five")
h.add(25, "twenty five")
h.add(26, "twenty six updated")
h.add(887, "large number")

print(h)
```

```python
# handle the hash collision
def add(self, key, value):
    key_hash = self._get_hash(key)
    key_value = [key, value]
    
    if self.map[key_hash] is None:
        self.map[key_hash] = [key_value]
        return True
    
    else:
        # check if it's an update
        for pair in self.map[key_hash]:
            if pair[0] == key:
                print("updating: ", key)
                pair[1] = value
                return True
            
        # nope, it's a collision: insert it
        self.map[key_hash].append(key_value)
        return True
HashMap.add = add
```

```python
h = HashMap()

h.add(17, "seventeen")
h.add(26, "twenty six")
h.add(35, "thirty five")
h.add(25, "twenty five")
h.add(26, "twenty six updated")
h.add(887, "large number")

print(h)
```

```python
def delete(self, key):
    key_hash = self._get_hash(key)
    
    if self.map[key_hash] is None:
        raise KeyError(str(key))
     
    for i in range(0, len(self.map[key_hash])):
        if self.map[key_hash][i][0] == key:
            self.map[key_hash].pop(i)
            return True
        
        
    # If the loop completes without finding the key
    raise KeyError(f"Key {key} not found")
HashMap.delete = delete
```

```python
h.delete(17)
h.delete(887)
print(h)
```

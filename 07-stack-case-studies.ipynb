# Writing Our Own

```python
class Stack:
    def __init__(self):
        self.l = []
        
    def push(self, val):
        self.l.append(val)
        
    def pop(self):
        return self.l.pop()
    
    def peek(self):
        return self.l[-1]
```

```python
s = Stack()
s.push(1)
s.push(2)
s.push(3)
# print(s[1])     # <-- error ... which, for us, is success since we have a stack and that's what we are looking for

# print(s)        # it will not give whole list
# print(s.pop())
```

```python
print(s.peek())
print(s.pop())
print(s.peek())
print(s.pop())
```

```python
a = [1,2,3]
# or a = '123'
b = [4,5,6]
# or b = '456'
print(list(zip(a,b)))
print(dict(zip(a,b)))
```

```python
def is_matched(string):
    opening = '({['
    closing = ')}]'
    mapping = dict(zip(opening, closing))

    stack = []
    
    for c in string:
        # case 1 : it's not opening bracket not closing so move to next bracket
        if c not in mapping.values() and c not in mapping.keys():
            continue
        
        # case 2
        # automatically checks only starting brackets
        if c in mapping:
            stack.append(mapping[c])    # we'll be looking for corresponding closing bracket
            
        # case 3: has to be closing bracket if we get here
        elif len(stack) == 0 or c!= stack.pop():
            return False
        
    return len(stack) == 0          # it checks if that there some bracket left in stack which is not close and it give false and is a mis match otherwise stack is empty and give true
```

```python
string = "{[[]]{()}}"
is_matched(string)
```

```python
string = "2 + (3 * 5) * ((2 * 2) * 5)"
is_matched(string)
```

```python
string = "2 + (3 * 5) * ((2 * 2) * 5) )"
is_matched(string)
```

# **Case Study: Binary to Decimal Conversion**

```python
def dec_to_bin(num):
    s1 = []
    
    if num < 0: 
        print("Negative value")
        return
    if num ==0:
        return num
    while num != 0:
        reminder = num % 2  # reminder will always be 0 or 1
        num = num // 2      # division by 2 gets rid of the least size of 2
        s1.append(reminder)
        
    ret = ''
    while s1:   #converting stack into string
        ret += str(s1.pop())    # comes out in reverse order of course
        
    return ret
```

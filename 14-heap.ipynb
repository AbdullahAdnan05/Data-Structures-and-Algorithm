# **Heap Data Structure and Heap Sort**

**Heap Data Structure - MaxHeap and MinHeap**

```python
def heapify(lst, n, root):
    """Heapify the root element of lst which has n elements in total. """
    
    largest = root
    l = 2*root + 1          # index of left hand side of given node
    r = 2*root + 2          # index of right hand side of given node
    
    if l < n and lst[l] > lst[largest]:         # n is length of list where l<n check if left side exsist
        largest = l
    if r < n and lst[r] > lst[largest]:         # n is length of list
        largest = r
        
    if largest != root:
        lst[root], lst[largest] = lst[largest], lst[root]       # swapping value
        
        heapify(lst, n, largest)        # recursive call
        
```

```python
def build_heap(lst):
    """Construct heap from any list by repeatedly applying heapify from bottom non-leaf to root"""
    n = len(lst)
    
    for i in reversed(range(n//2)):     # rest must be leaves, so, no need to heapify all the tree
        # print("heapifying: ", lst[i])
        heapify(lst, n, i)
```

```python
heap = [100, 5, 3, 2, 8, 15, 6, 102]
```

```python
print(heap)
build_heap(heap)
print(heap)
```

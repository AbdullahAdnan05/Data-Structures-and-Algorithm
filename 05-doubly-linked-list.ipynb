# Doubly Linked List
A doubly-linked list is quite similar to "singly-linked" list. Just to add a perv pointer to all the nodes

```python
class Node:
    def __init__(self, data=None):
        self.val = data
        self.next = None
        self.prev = None    #(change)
        
class Doubly:
    def __init__(self):
        self.head = None
```

**Push**

The push operation is the same mostly. the new node only needs to be linked to the second-to-last node.

```python
def push(self, val):
    new_node = Node(val)
    
    # no node currently 
    if self.head is None:
        self.head = new_node
        return

    # otherwise, reach the end and then insert
    last = self.head
    while last.next is not None:
        last = last.next 
        
    last.next = new_node
    new_node.prev = last        # Change
    
Doubly.push = push       ## We can add functiona to class even after definition
```

```python
def __str__(self):
    ret_str = '['
    temp = self.head
    while temp is not None:         # or just while temp
        # print(hex(id(temp.prev)), hex(id(temp)), hex(id(temp.next)))      # to check by id the link between nodes
        
        ret_str += str(temp.val)+ ', '
        temp = temp.next
        
    ret_str = ret_str.rstrip(', ')
    ret_str += ']'
    
    return ret_str
    
Doubly.__str__ = __str__
        
        
```

**POP**

```python
def pop(self):
    if self.head is None:
        raise Exception("cannot pop. No value ")
    
    # case where there is only one node
    if self.head.next is None:
        val = self.head.val
        self.head = None    # automatically gaarbage collection
        return val
    
    # case where there is 2 or more nodes
    # reach the previous to last node
    last = self.head
    while last.next is not None:
        prev = last
        last = last.next
    
    val = last.val
    prev.next = None  
    return val
    
Doubly.pop = pop      
     
```

**INSERT**

```python
def insert(self, index , val):
    new_node = Node(val)
    
    # insertion at index 0 is different
    if index == 0:
        new_node.next = self.head
        
        if self.head is not None:   # (change: check if head exsist )
            self.head.prev = new_node
        
        self.head = new_node
        
        return
    
    # insertion at any index
    temp = self.head
    
    counter = 0
    while  temp is not None and counter < index:
        pre = temp 
        temp = temp.next
        counter += 1
        # print(counter)
    
    # print("Will insert after: " )
    pre.next = new_node
    new_node.prev = pre     # (change)
    
    new_node.next = temp
    
    if temp.prev is not None:
        temp.prev = new_node    # (change)
    
Doubly.insert = insert
```

```python
l = Doubly()
l.push(1)
l.push(3)
l.insert(0,10)
l.insert(1,11)
print(l)
```

**REMOVE**

```python
def remove(self, val):
    
    temp = self.head
    
    # In Case 1
    if(val == temp.val):
        self.head = self.head.next
        if self.head is not None:
            self.head.prev = None
        return
    
    # In Case 2
    while temp is not None:
        if (temp.val == val):
            break
        
        pre = temp
        temp = temp.next
    
    if temp is None:
        return  # (Value not found)
    
    pre.next = temp.next
    if pre.next is not None:
        pre.next.prev = pre
                
    
    
Doubly.remove = remove
```

```python
l = Doubly()
l.push(1)
l.push(2)
l.push(3)
l.remove(2)
print(l)

l.remove(2)
print(l)

l.remove(1)
print(l)
```

**LENGTH**

```python
def len(self):
    temp = self.head
    counter = 1 
    while temp.next is not None:
        temp =  temp.next
        counter += 1
    
    return counter

Doubly.len = len
```

```python
l = Doubly()
l.push(1)
l.push(2)
l.push(3)
print(l)
l.len()
```

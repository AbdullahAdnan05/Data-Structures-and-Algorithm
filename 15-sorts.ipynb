# **Bubble Sort**

```python
def bubble_sort(l):
    n = len(l)
    # print(n)
    
    # Outer loop. Goes over the whole thing 'n' times
    # (because each time, one 'highest' will have moved to the end)
    for i in range(n):
        swapped = False
        # try to bubble the highest one up
        for j in range(0, (n-i)-1):
            
            # compare pairs, move higher one up (the highest will always reach the end )
            if l[j] > l[j+1]:
                l[j], l[j+1] = l[j+1], l[j]
                swapped = True
                
        if not swapped:
            break
```

```python
l = [1,2,4,1,2,5,5,6,1,110,15]
bubble_sort(l)
print(l)
```

# **Insertion Sort**

```python
def insertion_sort(l):
    
    # Go through all elements (except first).
    # Call it
    # Each time, key would be 'inserted' in its place
    # At each iteration, stuff less then i would be sorted already
    
    for i in range(1, len(l)):
        
        key = l[i]      # hold this key
        
        # start computing keys to things on its left!
        # stop when less or equal value found (or we reach left end)
        
        j = i-1
        while j >= 0 and key < l[j]:
            l[j+1] = l[j]       # move this to right. Slot left on j
            j -= 1
            
        l[j+1] = key        # Place key in free slot .....(j+1 because we decrease j above)
    
```

```python
l = [1, 2, 4, 1, 2, 3, 5, 5, 6, 1, 110, 19]
insertion_sort(l)
print(l)
```

# **Selection Sort**

```python
def selection_sort(l):
    n = len(l)
    
    # for each element in the list (starting from left)
    for i in range(n):
        min_idx = i     # find the minimun ...
        
        # ...... in the *rest* of the list
        for j in range(i+1, n):
            
            if l[j] < l[min_idx]:
                min_idx = j
                
        # swap the minimun with current element, now we have (sorted stuff till i)
        l[i], l[min_idx] = l[min_idx], l[i]
```

```python
l = [1, 2, 4, 1, 2, 5, 5, 6, 1, 110, 15]
selection_sort(l)
print(l)
```

# **Quick Sort**

```python
import random

def qsort(l, fst, lst):
    if fst >= lst: return
    
    i, j = fst, lst
    pivot = l[random.randint(fst, lst)]
    
    while i <= j:
        while l[i] < pivot: i +=1
        while l[j] > pivot: j -=1
        if i <= j:          # to swap only when i and j are not interchanged
            l[i], l[j] = l[j], l[i]
            i, j = i + 1, j - 1
    qsort(l, fst, j)
    qsort(l, i, lst) 
```

```python
l = [1.8, 2, 4, 1, 2, 5, 5, 6, 1, 110, 15]
qsort(l, 0, len(l)-1)
```

```python
print(l)
```

# **Sorting in the Real World**

# *Descending Sorts*

```python
def selection_sort(l):
    n = len(l)
    
    # for each element in the list (starting from left)
    for i in range(n):
        idx = i     # find the replacement ....
        
        # ..... in the *rest* of th list
        for j in range(i+1, n):
            
            if l[j] > l[idx]:       # changed to > for descending
                idx = j
                
        # swap the replacement with current element, now we have (sorted stuff till i)
        l[i], l[idx] = l[idx], l[i]
```

```python
l = [1, 2, 4, 1, 2, 5, 5, 6, 1, 110, 15]
selection_sort(l)
print(l)
```

But there's a better way!

```python
def less_than(a, b):
    return a < b
```

```python
def selection_sort(l, compare_with = lambda x, y: x < y):           # "lambda x, y: x < y" is a defualt function if not call by user
    n = len(l)
    
    # for each element in the list (starting from left)
    for i in range(n):
        min_idx = i        # find the minimum ....
        
        # ..... in the *rest* of the list
        for j in range(i+1, n):
            
            if compare_with(l[j], l[min_idx]):      # now the "comparison" is out-source
                min_idx = j
                
        # swap the minimum with current element, now we have (sorted stuff till i)
        l[i], l[min_idx] = l[min_idx], l[i]
```

```python
l = [1, 2, 4, 1, 2, 5, 5, 6, 1, 110, 15]
selection_sort(l, less_than)
print(l)
```

```python
def greater_than(a, b):
    return a > b
```

```python
l = [1, 2, 4, 1, 2, 5, 5, 6, 1, 110, 15]
selection_sort(l)       # it will do ascending sort by defualt
print(l)
selection_sort(l, greater_than)
print(l)
```

# **Taking it Even Further**

Now we can do all sorts of stuff with this without making a single change to our selection sort code.

```python
all_tuples = [ (24, 25),
               (1, 2),
               (2, 4),
               (3, 5)]
```

```python
def tuple_less_than(a, b):
    return sum(a) < sum(b)
def tuple_greater_than(a, b):
    return sum(a) > sum(b)
```

```python
print("Asending:\t", end="")
selection_sort(all_tuples, tuple_less_than)
print(all_tuples)

print("Desending:\t", end="")
selection_sort(all_tuples, tuple_greater_than)
print(all_tuples)
```

**Sorting in Python**

If you have a list of dictionaries -- each representing a student, for instance.

```python
d =  [
        {'name': 'Khalid',  'age':5},
        {'name': 'Usman',  'age':7},
        {'name': 'Ali',  'age':12},
        {'name': 'Farooq',  'age':3},
]
```

```python
def d_less_than(a, b):
    return a['age'] < b['age']

selection_sort(d, d_less_than)
print(d)
```

the above and the below are working same but in below we are making a key by which sorting can be done by using python build-in function

```python
def student_age(a):
    return a['age']         #  we are making age the key
def student_name(a):
    return a['name']        # we are making name the key 
# key are use for comparsion 
```

```python
print(d)
d.sort(key=student_age)         # sorting by age 
# the "key=student_age)" is sending the object to function and function is returning age from that object 
print(f"sorting by age ={d}")
d.sort(key=student_name)        # sorting by name
print(f"sorting by age ={d}")
```

by using it give ascending sort by default but if you what to do decending you have to do the following

```python
print(d)
d.sort(key=student_age, reverse=True)         # reverse=True tell it have to decending sort
print(f"sorting by age ={d}")
d.sort(key=student_name, reverse=True)        # sorting by name
print(f"sorting by age ={d}")
```

# **Sorting Objects of Custom Classes**

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return self.name + ': ' + str(self.age)
```

```python
s1 = Student('Wajid', 5)
s2 = Student('Usman', 2)
s3 = Student('Ali', 3)

s = [s1, s2, s3]
```
